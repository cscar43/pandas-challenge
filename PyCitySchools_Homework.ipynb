```python
#imports the need modules
import pandas as pd
import numpy as np

# Load data from CSV files
school_data= "./Resources/schools_complete.csv"
student_data= "./Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas
schools_df = pd.read_csv(school_data)
students_df = pd.read_csv(student_data)

# Combine the data
school_data_complete = pd.merge(schools_df, students_df, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

```python
#renames the column for type to school_type
school_data_complete = school_data_complete.rename(columns={'type': 'school_type'})
school_data_complete.head()
```

```python
# Calculate the total number of unique schools
total_schools = schools_df['school_name'].nunique()
```

```python
# Calculate the total number of students
total_students = students_df['student_name'].count()
```

```python
# Calculate the total budget
total_budget = schools_df['budget'].sum()
```

```python
# Calculate the average math score
avg_math_score = students_df['math_score'].mean()
```

```python
# Calculate the average reading score
avg_reading_score = students_df['reading_score'].mean()
```

```python
# Calculate the percentage of students passing math
passing_math = (students_df['math_score'] >= 70).sum() / total_students * 100
```

```python
# Calculate the percentage of students passing reading
passing_reading = (students_df['reading_score'] >= 70).sum() / total_students * 100
```

```python
# Calculate the percentage of students passing overall
passing_overall = (students_df[(students_df['math_score'] >= 70) & (students_df['reading_score'] >= 70)]['student_name'].count() / total_students) * 100
```

```python
# Create a new DataFrame to hold the district summary
district_summary_df = pd.DataFrame({
    'Total Schools': [total_schools],
    'Total Students': [total_students],
    'Total Budget': [total_budget],
    'Average Math Score': [avg_math_score],
    'Average Reading Score': [avg_reading_score],
    '% Passing Math': [passing_math],
    '% Passing Reading': [passing_reading],
    '% Overall Passing': [passing_overall]})

district_summary_df.head()
```

```python
# Group the data by school name and type
school_data = school_data_complete.groupby(['school_name', 'school type'])
```

```python
# Calculate the total number of students in each school
total_students = school_data['student_name'].count()
```

```python
# Calculate the total budget for each school
total_budget = school_data['budget'].mean()
```

```python
# Calculate the per student budget for each school
per_student_budget = total_budget / total_students
```

```python
# Calculate the average math score for each school
avg_math_score = school_data['math_score'].mean()
```

```python
# Calculate the average reading score for each school
avg_reading_score = school_data['reading_score'].mean()
```

```python
# Calculate the percentage of students passing math for each school
passing_math_percent = school_data_complete[school_data_complete['math_score'] >= 70].groupby('school_name')['math_score'].count() / total_students * 100
```

```python
# Calculate the percentage of students passing reading for each school
passing_reading_percent = school_data_complete[school_data_complete['reading_score'] >= 70].groupby('school_name')['reading_score'].count() / total_students * 100
```

```python
# Calculate the overall passing rate for each school
overall_passing_percent = (passing_math_percent + passing_reading_percent) / 2
```

```python
# Create a DataFrame to store the school summary data
school_summary_df = pd.DataFrame({
    'Total Students': total_students,
    'Total School Budget': total_budget,
    'Per Student Budget': per_student_budget,
    'Average Math Score': avg_math_score,
    'Average Reading Score': avg_reading_score,
    '% Passing Math': passing_math_percent,
    '% Passing Reading': passing_reading_percent,
    '% Overall Passing': overall_passing_percent
})
school_summary_df
```

```python
#creates a summary of the top 5 schools
top_schools = school_summary_df.sort_values("% Overall Passing", ascending=False).head(5)
top_schools.head()
```

```python
#creates a summary of the bottom 5 schools
bottom_schools = school_summary_df.sort_values("% Overall Passing").head(5)
bottom_schools.head()
```

```python
#shows the school and math scores by grade
math_scores_by_grade = pd.pivot_table(school_data_complete, values='math_score', index='school_name', columns='grade', aggfunc=np.mean)
math_scores_by_grade
```

```python
#shows the school and reading scores by grade
reading_scores_by_grade = pd.pivot_table(school_data_complete, values='reading_score', index='school_name', columns='grade', aggfunc=np.mean)
reading_scores_by_grade
```

```python
# create spending bins and labels
spending_bins = [0, 585, 630, 645, 680]
labels = ["<$585", "$585-630", "$630-645", "$645-680"]
```

```python
# create a new column "Spending Ranges (Per Student)" based on the bins and labels
school_summary_df["Spending Ranges (Per Student)"] = pd.cut(school_summary_df["Per Student Budget"], spending_bins, labels=labels)
school_summary_df
```

```python
# group by spending ranges and calculate the mean scores
spending_math_scores = school_summary_df.groupby(["Spending Ranges (Per Student)"]).mean()["Average Math Score"]
spending_reading_scores = school_summary_df.groupby(["Spending Ranges (Per Student)"]).mean()["Average Reading Score"]
spending_passing_math = school_summary_df.groupby(["Spending Ranges (Per Student)"]).mean()["% Passing Math"]
spending_passing_reading = school_summary_df.groupby(["Spending Ranges (Per Student)"]).mean()["% Passing Reading"]
overall_passing_spending = school_summary_df.groupby(["Spending Ranges (Per Student)"]).mean()["% Overall Passing"]
```

```python
# creates and shows the spending summary dataframe
spending_summary = pd.DataFrame({
    "Average Math Score": spending_math_scores,
    "Average Reading Score": spending_reading_scores,
    "% Passing Math": spending_passing_math,
    "% Passing Reading": spending_passing_reading,
    "% Overall Passing": overall_passing_spending
})
spending_summary
```

```python
# Bin the schools based on size
size_bins = [0, 1000, 2000, 5000]
size_labels = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
school_summary_df["School Size"] = pd.cut(school_summary_df["Total Students"], bins=size_bins, labels=size_labels)
school_summary_df
```

```python
# Calculate average scores and passing rates by school size
size_math_scores = school_summary_df.groupby("School Size").mean()["Average Math Score"]
size_reading_scores = school_summary_df.groupby("School Size").mean()["Average Reading Score"]
size_passing_math = school_summary_df.groupby("School Size").mean()["% Passing Math"]
size_passing_reading = school_summary_df.groupby("School Size").mean()["% Passing Reading"]
size_overall_passing = school_summary_df.groupby("School Size").mean()["% Overall Passing"]
```

```python
# Create a new DataFrame to hold the summary data
size_summary = pd.DataFrame({
    "Average Math Score": size_math_scores,
    "Average Reading Score": size_reading_scores,
    "% Passing Math": size_passing_math,
    "% Passing Reading": size_passing_reading,
    "% Overall Passing": size_overall_passing
})
size_summary
```

```python
# Group the per_school_summary DataFrame by school type
by_type = school_summary_df.groupby("school type")
```

```python
# Calculate the average scores and passing percentages by school type
type_math_scores = by_type.mean()["Average Math Score"]
type_reading_scores = by_type.mean()["Average Reading Score"]
type_passing_math = by_type.mean()["% Passing Math"]
type_passing_reading = by_type.mean()["% Passing Reading"]
type_overall_passing = by_type.mean()["% Overall Passing"]
```

```python
# Create a new DataFrame with the above calculations
type_summary = pd.DataFrame({
    "Average Math Score": type_math_scores,
    "Average Reading Score": type_reading_scores,
    "% Passing Math": type_passing_math,
    "% Passing Reading": type_passing_reading,
    "% Overall Passing": type_overall_passing
})
type_summary
```
